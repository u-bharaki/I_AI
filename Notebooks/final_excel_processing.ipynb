```python
import pandas as pd

# Son Ã§alÄ±ÅŸtÄ±ÄŸÄ±n dosyayÄ± kullan (gerekirse adÄ±nÄ± deÄŸiÅŸtir)
df = pd.read_csv(r"C:\Users\user\Desktop\cleaned_file_v15.csv")

def keep_first_part(x: str):
    if pd.isna(x):
        return x
    s = str(x)
    # TÃ¼m olasÄ± ayraÃ§larÄ± normal virgÃ¼le Ã§evir
    s = (s.replace('Ã¯Â¼Å’', ',')
           .replace('ï¼Œ', ',')
           .replace('ã€', ',')
           .replace(';', ',')
           .replace('|', ',')
           .replace('/', ','))
    # Ä°lk parÃ§ayÄ± al, kenar boÅŸluklarÄ±nÄ± temizle
    return s.split(',')[0].strip()

for col in ['Left-Diagnostic Keywords', 'Right-Diagnostic Keywords']:
    df[col] = df[col].apply(keep_first_part)

df.to_csv(r"C:\Users\user\Desktop\cleaned_file_v16.csv", index=False)
print("Bitti -> cleaned_file_v16.csv")
```

```python
import pandas as pd

# DosyayÄ± oku
df = pd.read_csv(r"C:\Users\user\Desktop\cleaned_file_v16.csv")

# --- 1ï¸âƒ£: Dosya yolu sÃ¼tununu otomatik bul ---
possible_path_cols = [c for c in df.columns if 'file' in c.lower() or 'path' in c.lower()]
if not possible_path_cols:
    raise ValueError("Filepath sÃ¼tunu bulunamadÄ±, lÃ¼tfen sÃ¼tun adÄ±nÄ± kontrol et!")
path_col = possible_path_cols[0]
print(f"KullanÄ±lan sÃ¼tun: {path_col}")

# --- 2ï¸âƒ£: Yeni 'Diagnosis' sÃ¼tununu oluÅŸtur ---
def select_diagnosis(row):
    filepath = str(row.get(path_col, '')).lower()
    if "_left" in filepath:
        return row.get('Left-Diagnostic Keywords')
    elif "_right" in filepath:
        return row.get('Right-Diagnostic Keywords')
    else:
        return None

df['Diagnosis'] = df.apply(select_diagnosis, axis=1)

# --- 3ï¸âƒ£: BoÅŸ kalan satÄ±r sayÄ±sÄ±nÄ± gÃ¶ster ---
empty_count = df['Diagnosis'].isna().sum()
print(f"BoÅŸ kalan Diagnosis satÄ±r sayÄ±sÄ±: {empty_count}")

# --- 4ï¸âƒ£: DosyayÄ± kaydet (Left/Right sÃ¼tunlarÄ±nÄ± korur) ---
df.to_csv(r"C:\Users\user\Desktop\cleaned_file_v17.csv", index=False)

print("âœ… TamamlandÄ±: Diagnosis sÃ¼tunu oluÅŸturuldu ve cleaned_file_v17.csv olarak kaydedildi.")
```

```python
import pandas as pd

# --- DosyayÄ± oku ---
DOSYA_ADI = r"C:\Users\user\Desktop\cleaned_file_final.csv"
df = pd.read_csv(DOSYA_ADI)

print(f"'{DOSYA_ADI}' baÅŸarÄ±yla okundu. Toplam {len(df)} satÄ±r veri bulundu.")
print("---" * 15)

# --- Cinsiyet DaÄŸÄ±lÄ±mÄ± ---
if 'Patient Sex' in df.columns:
    cinsiyet_dagilimi = df['Patient Sex'].value_counts(normalize=True) * 100
    print(" 1. Cinsiyet DaÄŸÄ±lÄ±mÄ±")
    for cinsiyet, oran in cinsiyet_dagilimi.items():
        print(f"  - {cinsiyet}: %{oran:.2f}")
    print("---" * 15)

# --- YaÅŸ Ä°statistikleri ---
if 'Patient Age' in df.columns:
    df['Patient Age'] = pd.to_numeric(df['Patient Age'], errors='coerce')
    print("2. YaÅŸ Ä°statistikleri")
    print(f"  - Minimum YaÅŸ: {df['Patient Age'].min()}")
    print(f"  - Maksimum YaÅŸ: {df['Patient Age'].max()}")
    print(f"  - Ortalama YaÅŸ: {df['Patient Age'].mean():.2f}")
    print("---" * 15)

# --- Diagnosis Etiket DaÄŸÄ±lÄ±mÄ± ---
print("3. HASTALIK DAÄILIMI (Diagnosis sÃ¼tununa gÃ¶re)")
if 'Diagnosis' in df.columns:
    diagnosis_counts = df['Diagnosis'].value_counts()
    diagnosis_percentages = (diagnosis_counts / len(df)) * 100

    print(f"Toplam {len(df)} kayÄ±t Ã¼zerinden yÃ¼zdesel daÄŸÄ±lÄ±m:\n")
    for diag, count in diagnosis_counts.items():
        print(f"  - {diag:<12}: {count} kayÄ±t (%{diagnosis_percentages[diag]:.2f})")
else:
    print("'Diagnosis' sÃ¼tunu bulunamadÄ±.")
print("---" * 15)

print("TÃ¼m istatistikler baÅŸarÄ±yla oluÅŸturuldu.")
```
